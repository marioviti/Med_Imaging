```python
def thrs(t,img):
    shape_ = img.shape
    nuimg = np.array(map(lambda x : 1 if x>t else 0,img.reshape(-1))).reshape(shape_)
    return nuimg

images_name_case = []
images_meta_data = []
images_data = []
masks_data = []
#for img_name in sorted(utils.getdirlist(nii_dir,extensions)):
for img_name in sorted(settings['todo_dirlist']):
    print img_name
    images_name_case += [img_name]
    image_data_and_metadata = (nio.basicLoadNii(utils.joinpath(nii_dir,img_name)))
    image = image_data_and_metadata[2]
    print image_data_and_metadata[0]
    images_meta_data += [(image_data_and_metadata[0],image_data_and_metadata[1])]
    images_data += [image]
    axial = image[int(image.shape[0]/2),:,:] # RL
    coronal = image[:,int(image.shape[1]/2),:] # AP
    saggittal = image[:,:,int(image.shape[2]/2)] # SI
    plt.imshow(axial)
    print 'axial'
    plt.show()
    plt.imshow(coronal)
    print 'coronal'
    plt.show()
    plt.imshow(saggittal)
    print 'saggittal'
    plt.show()
    tissues = []
    for seg_img_name in sorted(utils.getdirlist(nii_segm_dir,extensions)):
        if img_name in seg_img_name:
            print seg_img_name
            tissue = np.array(nio.basicLoadNii(utils.joinpath(nii_segm_dir,seg_img_name)))[2]
            axial = tissue[int(tissue.shape[0]/2),:,:] # RL
            coronal = tissue[:,int(tissue.shape[1]/2),:] # AP
            saggittal = tissue[:,:,int(tissue.shape[2]/2)] # SI
            plt.imshow(thrs(0.7,axial))
            print 'axial'
            plt.show()
            plt.imshow(thrs(0.7,coronal))
            print 'coronal'
            plt.show()
            plt.imshow(thrs(0.7,saggittal))
            print 'saggittal'
            plt.show()
            tissues += [tissue]
    masks_data += [tissues]
    print len(tissues)
        
```

```python
tissues_data = [] # 3 arrays containing grey values corresponding to areas in masks

print 'masking: this could take time'
i = -1
for img in images_data:
    i+=1
    img_unroll = img.reshape(-1)
    tissues = []
    for mask in masks_data[i]:
        hist = []  
        mask_unroll = thrs(0.7,mask.reshape(-1))
        tissue = img_unroll * mask_unroll
        tissue = [ voxel for voxel in tissue if voxel > 0 ]
        print len(tissue)
        tissues += [tissue]
    tissues_data += [tissues]
print 'masking: finished'
```

# Reverse engeneerig the segmentation

We assume we have the best threshold values for downsampling.
This thresholds are extracted from sampling values within the masked regions of each tissue.

```python
for tissues in tissues_data:
    i = 0
    tissue = []
    for tissue in tissues:
        if i == 0:
            plt.hist(tissue, bins=128, range=[0.0, 1.5], color='green', alpha=0.3)
        if i == 1:
            plt.hist(tissue, bins=128, range=[0.0, 1.5], color='red', alpha=0.3)
        if i == 2:
            plt.hist(tissue, bins=128, range=[0.0, 1.5], color='blue', alpha=0.3)
        i+=1
    plt.show()
   
```

we notice a mixture of 3 gaussians as expected.

# Recompute EM on single gaussian
In order to get the most representative value for the threshold we recompute the EM estimation from each histogram.

```python
parameters_data = []
g = GMM(n_components=1,verbose=1)
bins_ = 128
for tissues in tissues_data:
    parameters = []
    for tissue in tissues:
        y,x = np.histogram(tissue, bins=bins_, density=True)
        x = x[0:bins_]
        g.fit(np.array(tissue).reshape(-1,1))
        parameters += [g.means_[0]]
        density = np.exp(g.score(x.reshape(-1,1)))
        plt.plot(x[0:bins_], y[0:bins_], '-r')
        plt.plot(x, density, '-b')
        plt.show()
        print g.means_
    parameters_data += [sorted(parameters)]
```

```python
print parameters_data
```

now that we have our parameters we'll evenly space the intervals to get the thresholds

```python
thress_data = []
min_cut_grey,max_cut_grey =(0.050,1.5)
i = -1
for img in images_data:
    i+=1
    img_data_voxel = img.reshape(-1)
    plt.hist(img_data_voxel, bins=bins_, color='blue', lw=2)
    thress = []
    for parameters in parameters_data[i]:
        plt.vlines(parameters, 0, -100000, color='red')
    thress = []
    for j in range(0,len(parameters_data[i])):
        if j == 0:
            thress += [ (min_cut_grey+parameters_data[i][j][0])/float(2) ]
        elif j == len(parameters_data[i])-1 :
            thress += [ (parameters_data[i][j-1][0]+parameters_data[i][j][0])/float(2) ]
            thress += [ (parameters_data[i][j][0]+max_cut_grey)/float(2) ]
        else :
            thress += [ (parameters_data[i][j-1][0]+parameters_data[i][j][0])/float(2) ]
    for thr in thress:
        print thr
        plt.vlines(thr, 0, -101000, color='green')
    thress = sorted(thress)
    thress_data += [thress]
    plt.show()
```

```python
def segment(min_x,max_x,img):
    shape_ = img.shape
    img = img.reshape(-1)
    img = np.array(map(lambda x: 1 if x>=min_x and x<max_x else 0, img))
    return img.reshape(shape_)

encoded_images = []
c_N_data = []
for img in images_data:
    c_N = []
    c_N += [segment(thress[0],thress[1],img)]
    c_N += [segment(thress[1],thress[2],img)]
    c_N += [segment(thress[2],thress[3],img)]
    c_N_data += [c_N]
    encoded_img = c_N[0]*0.33+c_N[1]*0.66+c_N[2]
    encoded_images += [encoded_img]
    for c_i in c_N:
        axial = c_i[int(c_i.shape[0]/2),:,:] # RL
        coronal = c_i[:,int(c_i.shape[1]/2),:] # AP
        saggittal = c_i[:,:,int(c_i.shape[2]/4)] # SI
        plt.imshow(axial)
        plt.show()
        plt.imshow(coronal)
        plt.show()
        plt.imshow(saggittal)
        plt.show()
    plt.imshow(encoded_img[int(c_i.shape[0]/2),:,:])
    plt.show()
```
